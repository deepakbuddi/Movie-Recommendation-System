```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
sel = ['genres','keywords','cast','director']
print(sel)
```

```python
for i in sel:
  movies_data[i] = movies_data[i].fillna('')
```

```python
combined = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['cast']+' '+movies_data['director']

print(combined)
```

```python
vec = TfidfVectorizer()
featured = vec.fit_transform(combined)
print(featured)
```

```python
similarity = cosine_similarity(featured)
print(similarity)
```

```python
similarity.shape
```

```python
list_of_titles = movies_data['title'].tolist()
print(list_of_titles)
```

```python
name = input("Enter Movie Name: ")
```

```python
close_match = difflib.get_close_matches(name, list_of_titles)
print(close_match)
```

```python
c_match = close_match[0]
print(c_match)
```

```python
index = movies_data[movies_data.title == c_match]['index'].values[0]
print(index)
```

```python
similarity_score = list(enumerate(similarity[index]))
print(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  idx = movie[0]
  title_from_index = movies_data[movies_data.index==idx]['title'].values[0]
  if (i<=5):
    print(i, '.',title_from_index)
    i+=1
```

```python
movie_name = input(' Enter movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<6):
    print(i, '.',title_from_index)
    i+=1
```
